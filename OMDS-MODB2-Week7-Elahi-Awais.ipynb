# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

To pull the viewer to the last three months first, I would change those points or line segments so they stand out immediately from the rest of the chart. The cleanest option is color: keep the first nine months in a muted gray and use a saturated accent color for the final three months.

Other strong preattentive options are size and weight, such as thicker line segments, larger markers, or bold endpoint labels for the last three months only. I could also use enclosure by lightly shading the final quarter, or add a direct annotation near the last three points so the eye lands there first. A final option is intensity or contrast: reduce the opacity of the earlier months and keep the last three months fully opaque.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
highlight_df = pd.DataFrame({"Date": datearray, "Phosphate": time_series})
base = highlight_df.iloc[:-3]
focus = highlight_df.iloc[-3:]

ax = sns.lineplot(data = base, x = "Date", y = "Phosphate", color = "#bdbdbd", linewidth = 2)
sns.lineplot(data = focus, x = "Date", y = "Phosphate", color = "#d95f02", linewidth = 4, marker = "o", ax = ax)
ax.axvspan(focus["Date"].iloc[0], focus["Date"].iloc[-1], color = "#fdd0a2", alpha = 0.25)
ax.set_title("Highlighting the Final Three Months")
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

comparison_df = pd.DataFrame({
    "metric": ["Awareness", "Interest", "Consideration", "Trial", "Repeat purchase"],
    "before": [62, 57, 49, 31, 18],
    "after": [62, 60, 56, 42, 29]
})

fig, ax = plt.subplots(figsize = (8, 4.5))
for _, row in comparison_df.iterrows():
    ax.plot([row["before"], row["after"]], [row["metric"], row["metric"]], color = "#bdbdbd", linewidth = 2)
    ax.scatter(row["before"], row["metric"], color = "#9e9e9e", s = 50)
    ax.scatter(row["after"], row["metric"], color = "#1b9e77", s = 70)

ax.set_title("Brand Funnel Improvement After Campaign", loc = "left")
ax.set_xlabel("Percent of customers")
ax.spines[["top", "right", "left"]].set_visible(False)
ax.tick_params(axis = "y", length = 0)
ax.grid(axis = "x", alpha = 0.2)
ax.annotate(
    "Largest gain appears in trial",
    xy = (42, "Trial"),
    xytext = (49, "Consideration"),
    arrowprops = dict(arrowstyle = "->", color = "#1b9e77"),
    color = "#1b9e77"
)
plt.tight_layout()
```
